```python
def search_text(text):
    results = es.search(index='data',query={"match_phrase":{"Body":text}})
    all_hits = results['hits']['hits']

    # see how many "hits" it returned using the len() function
    print ("total hits :", results['hits']['total']['value'], "\n")
    
    # iterate the nested dictionaries inside the ["hits"]["hits"] list and print results
    if results['hits']['total']['value'] == 0:
        print('NOT FOUND')  
    else:
        for num, doc in enumerate(all_hits):
            print ("News_ID:", doc['_source']["News_ID"], "\n", 'Title : ', doc['_source']['Title'],'\n' , 'Link : ' , doc['_source']['Short_link'],'\n'  , doc['_source']['Body'], "\n")
```

```python
def search_date(text):
    results = es.search(index='data',query={"match_phrase":{"Date":text}})
    all_hits = results['hits']['hits']

    # see how many "hits" it returned using the len() function
    print ("total hits :", results['hits']['total']['value'], "\n")
    
    # iterate the nested dictionaries inside the ["hits"]["hits"] list and print results
    if results['hits']['total']['value'] == 0:
        print('NOT FOUND')  
    else:
        for num, doc in enumerate(all_hits):
            print ("News ID:", int(doc['_source']['News_ID']), "\n")
```

```python
def search_correctedtext(text):
    tokens = text.split(" ")
    results = es.search(index='data',query={"bool": {"must": [{"match": {'Body':{"query": i,"fuzziness":1 }}} for i in tokens ]  }})
    all_hits = results['hits']['hits']

    # see how many "hits" it returned using the len() function
    print ("total hits :", results['hits']['total']['value'], "\n")
    
    # iterate the nested dictionaries inside the ["hits"]["hits"] list and print results
    if results['hits']['total']['value'] == 0:
        print('NOT FOUND')  
    else:
        for num, doc in enumerate(all_hits):
            print ("DOC ID:", doc["_id"], "\n", doc['_source']['Body'], "\n")
```
